```python
import pandas as pd
import altair as alt

df = pd.read_csv('set.csv')
productions = {}

for pros in df['Job']:
    if not(isinstance(pros, float) and pd.isna(pros)):
        if "Yes" in pros:
            pros = pros.replace("Yes, ", "")
        if "Big Tech" in pros:
            pros = pros.replace('Big Tech (IBM, JP Morgan, Bank of America, etc)', 'Big Tech')
        if 'No' in pros:
            pros = 'None'
        pro_list = pros.split(', ')
        for pro in pro_list:
            if pro in productions:
                productions[pro] += 1
            else:
                productions[pro] = 1

productions_df = pd.DataFrame(list(productions.items()), columns=['Company Bias', 'Count'])
chart = alt.Chart(productions_df).mark_arc().encode(
    theta='Count:Q',
    color='Company Bias:N',
    tooltip=['Style of Production:N', 'Count:Q']
).properties(
    width=800,
    height=600
)
chart
```
