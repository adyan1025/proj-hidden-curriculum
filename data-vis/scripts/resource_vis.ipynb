```python
import pandas as pd
import altair as alt

df = pd.read_csv('set.csv')
productions = {}

for pros in df['Resource']:
    if not(isinstance(pros, float) and pd.isna(pros)):
        if not('NA' in pros):
            if 'Networking' in pros:
                pros = pros.replace('Networking (Linkedin / Recruiter / Conferences)', 'Networking')
            if 'Community' in pros:
                pros = pros.replace('Online Community (follow me on Tiktok / Discord)', 'Online Community')
            pro_list = pros.split(', ')
            for pro in pro_list:
                if pro in productions:
                    productions[pro] += 1
                else:
                    productions[pro] = 1

productions_df = pd.DataFrame(list(productions.items()), columns=['Opportunity or Resource', 'Count'])
chart = alt.Chart(productions_df).mark_arc(innerRadius=50).encode(
    theta=alt.Theta('Count:Q', title='Count'),
    color=alt.Color('Opportunity or Resource:N', title='Opportunity or Resource')
).properties(
    title='Distribution of Opportunities and Resources',
    width = 800,
    height = 600
)
chart
```
