```python
import pandas as pd
import altair as alt

df = pd.read_csv('set.csv')
productions = {}

for pros in df['Perks']:
    if not(isinstance(pros, float) and pd.isna(pros)):
        if "Office Perks" in pros:
            pros = pros.replace('Office Perks (location, office activities, campus)', 'Office Perks')
        if "Enjoy food / entertainment at work":
            pros = pros.replace('Enjoy food / entertainment at work', 'Entertainment at Work')
        pro_list = pros.split(', ')
        for pro in pro_list:
            if pro in productions:
                productions[pro] += 1
            else:
                productions[pro] = 1

productions_df = pd.DataFrame(list(productions.items()), columns=['Why Work There?', 'values'])
base = alt.Chart(productions_df).encode(
    alt.Theta("values:Q").stack(True),
    alt.Radius("values:Q").scale(type="sqrt", zero=True, rangeMin=20),
    color="Why Work There?:N",
).properties(
    width = 800,
    height = 600
)
c1 = base.mark_arc(innerRadius=20, stroke="#fff")
c2 = base.mark_text(radiusOffset=10).encode(text="values:Q")
c1 + c2
```
