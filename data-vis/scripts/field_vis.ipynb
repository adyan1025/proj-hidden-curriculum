```python
import pandas as pd
import altair as alt
from altair_saver import save


df = pd.read_csv('set.csv')
productions = {}

for pros in df['Fields']:
    if not(isinstance(pros, float) and pd.isna(pros)):
        if "Product" in pros:
            pros = pros.replace('Product / Project Management', 'Project Management')
        if "General Computer Science" in pros:
            pros = pros.replace('General Computer Science', 'General CS')
        if "Quant" in pros:
            pros = pros.replace('Quant', 'Quantitative Analyst')
        pro_list = pros.split(', ')
        for pro in pro_list:
            if pro in productions:
                productions[pro] += 1
            else:
                productions[pro] = 1

productions_df = pd.DataFrame(list(productions.items()), columns=['CS Field', 'Count'])
productions_df = productions_df.sort_values(by='Count', ascending=True)

bars = alt.Chart(productions_df).mark_bar(color='#2e2e2e').encode(
    x=alt.X('Count:Q', title='Count'),
    y=alt.Y('CS Field:N', title='CS Field', sort=productions_df['CS Field'].tolist())
).properties(
    title='Distribution of CS Fields',
    width = 800,
    height = 600,
)
text = bars.mark_text(
    dx = 10,
    color='black'
).encode(
    text='Count:Q'
)

chart = bars + text
chart
```
